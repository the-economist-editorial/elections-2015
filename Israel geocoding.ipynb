```python
import requests
import json
```

```python
API_URL = 'http://maps.googleapis.com/maps/api/geocode/json'
def single_geocode(address):
    print('fetching %s' % address)
    res = requests.get(API_URL, params={
            'address' : address,
            'sensor' : 'false',
        })
    return json.loads(res.content.decode('utf8'))
    
def bulk_geocode(addresses):
#     res = {}
#     url = 'http://maps.googleapis.com/maps/api/geocode/json'
    return [{address : single_geocode(address)} for address in addresses]
#     for address in addresses:
#         # print 'fetching ' + address
#         res = requests.get(url, params={
#             'address': address,
#             'sensor': 'false'
#         })
# #         address = unicode(address.decode('utf8')).encode('utf8')
#         print(address)
#         res[address] = json.loads(res.content.decode('utf8'), encoding='utf8')
#     return res
```

```python
geodat_dict = {list(el.keys())[0]:list(el.values())[0] for el in geodat}
```

```python
def get_adm(results):
    return [area for area in results[0]['address_components'] if area['types'][0] == 'administrative_area_level_1'][0]['short_name']
```

```python
def get_adm_from_name(name):
    if not name in geodat_dict.keys():
        return None
    return get_adm(geodat_dict[name]['results'])
```
