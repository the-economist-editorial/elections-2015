```python
import requests
import json
```

```python
API_URL = 'http://maps.googleapis.com/maps/api/geocode/json'
def single_geocode(address):
    print('fetching %s' % address)
    res = requests.get(API_URL, params={
            'address' : address,
            'sensor' : 'false',
        })
    return json.loads(res.content.decode('utf8'))
    
def bulk_geocode(addresses):
    return {address : single_geocode(address) for address in addresses}
```

```python
def get_adm(results):
    if len(results) == 0:
        return None
    national = [area for area in results[0]['address_components'] if area['types'][0] == 'country']
    if len(national) == 0 or national[0]['short_name'] != 'IL':
        return None
    areas = [area for area in results[0]['address_components'] if area['types'][0] == 'administrative_area_level_1']
    if len(areas) == 0:
        return None
    return areas[0]['short_name']
```

```python
def get_adm_from_name(name):
#     print(name)
    if not name in geodat.keys():
        return None
    return get_adm(geodat[name]['results'])
```

```python
def dHondt(name, row, parties):
    seats = int(row['Seats'])
    votes = [row[p] for p in parties]
    denominators = range(1,seats+1)
    coeffs = [[{ 'party' : parties[idx], 'coeff' : v/d} for d in denominators] for idx,v in enumerate(votes)]
    ranked_list = sorted([item for sublist in coeffs for item in sublist], key=lambda x:x['coeff'])
    winners = pd.DataFrame(ranked_list[-seats:])
    results = winners.groupby('party').count().rename(columns={'coeff':'seats'}).reset_index()
    results['province'] = name
    return results
```
