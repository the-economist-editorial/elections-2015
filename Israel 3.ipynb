```python
import pandas as pd
import numpy as np
import json
```

```python
returns = pd.read_excel('results/israel/knesset-city.xlsx')
```

```python
with open('results/israel/geocoded.json') as f:
    geodat = json.loads(f.read())
```

```python
geodat
```

```python
def get_adm(results):
    if len(results) == 0:
        return 'No result'
    areas = [area for area in results[0]['address_components'] if len(area['types']) > 0 and area['types'][0] == 'administrative_area_level_1']
    if len(areas) == 0:
        return 'Not in district'
    val = areas[0]['short_name']
    if val == 'מחוז הדרום':
        return 'South District'
    if val == 'מחוז הצפון':
        return 'North District'
    return val
def get_adm_from_name(name):
    if not name in geodat.keys():
        return None
    return get_adm(geodat[name]['results'])
```

```python
def get_loc(results):
    if len(results) == 0:
        return 'NA'
    geom = results[0]['geometry']
    return geom['location']
def get_lat_from_name(name):
    if not name in geodat.keys():
        return None
    loc = get_loc(geodat[name]['results'])
    if loc == 'NA':
        return None
    return loc['lat']
def get_lng_from_name(name):
    if not name in geodat.keys():
        return None
    loc = get_loc(geodat[name]['results'])
    if loc == 'NA':
        return None
    return loc['lng']
```

```python
returns['district'] = returns['Name'].map(get_adm_from_name)
returns['latitude'] = returns['Name'].map(get_lat_from_name)
returns['longitude'] = returns['Name'].map(get_lng_from_name)
returns.head()
```

```python
returns.groupby('district').sum()
```

```python
returns['subdistricts'] = returns['Name'].map(get_subadm_from_name)
```
