```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as srs
from __future__ import division
```

```python
uk = pd.read_excel('./results/uk-2015.xls', sheetname='Results')
uk.head()
```

```python
uk_sums = uk.groupby('Region')['Turnout', 'Conservative_Party', 'Labour_Party', 'Liberal_Democrats', 'Green', 'UK_Independence_Party', 'Scottish_National_Party', 'Plaid_Cymru', 'Respect_Unity_Coalition', 'Sinn_Fein', 'SDLP', 'DUP', 'Ulster_Unionist','Alliance'].sum().fillna(0)
```

```python
uk_sums.div(uk_sums.Turnout, axis='index')
```

```python
seats = uk.groupby('Region')['Name'].count()
```

```python
combine = pd.concat([seats, uk_sums], axis='index').rename(columns={'Name' : 'Seats'})
combine
```

```python
parties = ['Conservative_Party', 'Labour_Party', 'Liberal_Democrats', 'Green', 'UK_Independence_Party', 'Scottish_National_Party', 'Plaid_Cymru', 'Respect_Unity_Coalition', 'Sinn_Fein', 'SDLP', 'DUP', 'Ulster_Unionist', 'Alliance']
```

```python
[series['Seats'] for name, series in combine.iterrows()]
```

```python
def dHondt(name, row, parties):
    seats = int(row['Seats'])
    votes = [row[p] for p in parties]
    denominators = range(1,seats+1)
    coeffs = [[{ 'party' : parties[idx], 'coeff' : v/d} for d in denominators] for idx,v in enumerate(votes)]
    ranked_list = sorted([item for sublist in coeffs for item in sublist], key=lambda x:x['coeff'])
    winners = pd.DataFrame(ranked_list[-seats:])
    results = winners.groupby('party').count().rename(columns={'coeff':'seats'}).reset_index()
    results['Region'] = name
    return results
```

```python
provinces = [dHondt(region, r, parties) for region, r in combine.iterrows()]
pd.concat(provinces).groupby('party').sum().sort('seats', ascending=False)
```

```python
party_total_votes = uk_sums[parties].sum()
total_votes = party_total_votes.sum()
party_pcts = party_total_votes / total_votes
```

```python
lim_parties = party_pcts[party_pcts > 0.1].index.tolist()
lim_provinces = [dHondt(region, r, lim_parties) for region, r in combine.iterrows()]
pd.concat(lim_provinces).groupby('party').sum()
```

```python
national = uk[parties].sum()
national['Seats'] = uk['Name'].count()
nat_prop = pd.DataFrame(dHondt('all', national, parties))
nat_prop.set_index(nat_prop.party).sort('seats',ascending=False)['seats']
```
